```python
import numpy as np
```

```python
map_np = np.asarray(data['map'])

h, w = map_np.shape
```

```python
canvas = [[' '] * w] * h
for yy in range(h):
    for xx in range(w):
        canvas[xx][yy] = str(map_np[xx, yy])
        
```

```python
def print_canvas(canvas):
    canvas_str = ''
    for ii in range(len(canvas)):
        canvas_str += ''.join(canvas[ii]) + '\n'
    print(canvas_str)
    
print_canvas(canvas)
```
